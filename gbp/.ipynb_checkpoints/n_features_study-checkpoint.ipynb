```python
Sigma_prior = 3

Sigma_fix = 0.0008

Sigma_measurement = 1
loss = 'huber'
Nstds = 5

meas_var = 1.

KL_threshold = -1
```

```python
def test(iters, n_frames, frac_map):
    
    data_dir = '../ba_data/TUM/rgbd_dataset_freiburg3_long_office_household'
    cam_params, timestamps, landmarks, pointIDs, features, feature_camIDs, \
                feature_pointIDs, n_points, n_cameras, cam_properties = import_data(data_dir, frac_map=frac_map)
    # cam_params, landmarks, cam_params_gt, landmarks_gt, n_cameras, n_points, timestamps, pointIDs, \
    #              features, feature_camIDs, feature_pointIDs, cam_properties = load_data('2M_2N')

    print(n_points)

    cam_params, landmarks = np.array(cam_params), np.array(landmarks)
    # np.random.seed(0)
    # cam_params[1:] += np.random.normal(0, 0.1, cam_params[1:].shape)
    # landmarks[:] += np.random.normal(0, 0.1, landmarks[:].shape)

    # Just use first 10 frames of sequence
    first = np.random.randint(len(cam_params) - 5)
    print(first)
    last = first + n_frames
    cam_params, timestamps, landmarks, pointIDs, n_cameras, n_points, features, feature_camIDs, feature_pointIDs = getFrameSequence(first, last, cam_params, \
                            timestamps, landmarks, pointIDs, n_cameras, n_points, features, feature_camIDs, feature_pointIDs)

    print('\nNumber of frames: ', n_cameras)
    print('Number of landmarks: ', n_points)

    cam_dofs = len(cam_params[0])
    print('Camera dofs: ', cam_dofs, '\n')

    n_obs_features = len(features)
    n_residuals = n_obs_features * 2
    n_vars = n_cameras * cam_dofs + n_points * 3
    print('Number of residuals:', n_residuals)
    print('Number of variables:', n_vars)
    if n_vars < n_residuals:
        print('Constrained', '\n')
    else:
        print('Unconstrained', '\n')

    for camID in range(n_cameras):
        ncam_features = feature_camIDs.count(camID)
        print(f'Frame {camID} observes {ncam_features} landmarks.')
        point_ix = [i for i,x in enumerate(feature_camIDs) if x==camID]
        pointIDs_obs = np.array(feature_pointIDs)[np.array(point_ix)]
        print(pointIDs_obs)

        common_observations = []
        for pointID in pointIDs_obs:
            cam_ix = [i for i,x in enumerate(feature_pointIDs) if x==pointID]
            common_observations += list(np.array(feature_camIDs)[np.array(cam_ix)])

        print('CamIDs of other cameras that observe at least one of the same map points: ', set(common_observations))

    cam_fix = [0]
    lmark_fix = []
    graph = createGraph(cam_params, landmarks, pointIDs, features, feature_camIDs, feature_pointIDs, cam_properties, 
                    Sigma_measurement, loss, Nstds, KL_threshold, meas_var, Sigma_fix, Sigma_prior, cam_fix, lmark_fix, timestamps=timestamps)



    graph = sendPriors(graph, lmark_prior=True)

    graph = update_varnodes(graph)

    print('Reprojection Error: ', av_error(graph))


    # Can increase damping after we have sent priors
    graph.prob_damping = 0.0  # probability damping factor
    graph.mean_damping = 0.0
    graph.message_damping = 0.95  # message damping factor

    err_log = []
    for i in range(iters):
        graph = flooding(graph)

        err = av_error(graph)
        err_log.append(err)
        print(f'{i} Reprojection error: ', err)

        n_messages_sent, frac_messages_sent = KL_analysis(graph)
        print(n_messages_sent, frac_messages_sent)

        
    points_per_frame = len([x for list in graph.meas_factor_nodes for x in list])   
    
    return err_log, points_per_frame
```
